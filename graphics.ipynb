```python
import pandas
import matplotlib.pyplot as plt
```

```python
df = pandas.read_csv('distances.csv')
```

```python
for col in df.columns[:-1]:

  data = df[col].dropna()
  
  data_min, data_max = min(data), max(data)
  bin_width = (data_max - data_min) / 10

  plt.figure(figsize=(8, 6))

  #plt.hist(data, bins=30, color='skyblue', edgecolor='black')
  plt.hist(data, bins=10, range=(data_min, data_max + bin_width), color='skyblue', edgecolor='black')

  plt.title(f'Histograma de las distancias entre puntos con dimension {col}')
  plt.xlabel('Distancia')
  plt.ylabel('Frecuencia')
  
  plt.xlim(0)
  
  plt.grid(True)
  plt.show()
```
